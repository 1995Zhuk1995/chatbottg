```python
from __future__ import print_function
import random
import codecs
import pandas as pd
import matplotlib.pyplot as plt
from Tokenizer import Tokenizer
%matplotlib inline
```

```python
df = pd.read_csv(u'../data/premise_question_answer.csv', encoding='utf-8', delimiter='\t', index_col=None)
```

```python
df.shape[0]
```

```python
tokenizer = Tokenizer()
df['answer_len'] = df['answer'].apply( func=lambda z:len(tokenizer.tokenize(z)) )
df['premise_len'] = df['premise'].apply( func=lambda z:len(tokenizer.tokenize(z)) )
df['question_len'] = df['question'].apply( func=lambda z:len(tokenizer.tokenize(z)) )
```

### Длины вопросов:

```python
for answer_len in sorted(df['answer_len'].unique()):
    n = sum(df['answer_len']==answer_len)
    print('len={} count={}'.format(answer_len, n))
```

### Длины предпосылок:

```python
for premise_len in sorted(df['premise_len'].unique()):
    n = sum(df['premise_len']==premise_len)
    print('len={} count={}'.format(premise_len, n))
```
